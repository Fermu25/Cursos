```python
#Método de monte Carlo, "Hit and miss" calculamos el área de un círculo unitario, encerramos el círculo en un cuadrado. Lanzamos N piedras y contamos N_m, A_cir=N_in/N *A_cuadrado

import random as r
import math as m
n_in=0
n=100000000 #!millón de pasos para acercarse a la segunda cifra decimal de pi
for i in range(0,n):
  x2=r.random()**2 #Puntos dentro del círculo, coordenadas polares y son puntos al azar
  y2=r.random()**2
  if m.sqrt(x2+y2)<1.0: #Esta condición marca que esté dentro del círculo, el radio,dentro de la circunferencia
    n_in += 1
L_cuadrado=2 #Definimos un cuadrado de lado 2, porque el círculo es de radio 1
area=(n_in)/ float(n)*L_cuadrado**2

print(area)
```

```python
import numpy as np
f=lambda x: m.sqrt(1-x**2)

def simpson(f, a, b, n):
    h=(b-a)/n
    s=0.0
    x=a + h
    for i in range(1,n,2):
      s+=4.0*f(x)
      x += 2*h

    x = a + 2*h
    for i in range(2,n,2):
        s += 2*f(x)
        x += 2*h
    return (h/3)*(f(a)+f(b)+s)
x=np.linspace(0,np.pi,0.1)
print(simpson(f(x),0,np.pi,1000))
#VEntaja puedo usar este método para cualquier  función, desventaja el tiempo de cómputo y la precisión
```
