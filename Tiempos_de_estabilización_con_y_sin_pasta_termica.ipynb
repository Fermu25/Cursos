```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# === Cargar archivo CSV ===
archivo = 'estabilizacio.csv'  # Reemplaza con el nombre real de tu archivo
df = pd.read_csv(archivo)

# === Procesar tiempos (Grupo 1 y Grupo 2) ===
def convertir_a_segundos(horas):
    base = datetime.strptime(horas.iloc[0], "%H:%M:%S.%f")
    return horas.apply(lambda x: (datetime.strptime(x, "%H:%M:%S.%f") - base).total_seconds())

# === Procesar grupo 1 (columnas 0 a 4) ===
df_g1 = df.iloc[:, 0:5].copy()
df_g1.columns = ['Hora', 'T1_TCOR', 'T2_TCOR', 'T3_TCOR', 'T4_TCOR']
df_g1['Hora'] = df_g1['Hora'].str.extract(r'(\d{2}:\d{2}:\d{2}\.\d{3})')  # Extrae hora
df_g1.dropna(inplace=True)
df_g1['Tiempo_s'] = convertir_a_segundos(df_g1['Hora'])

# === Procesar grupo 2 (columnas 5 a 9) ===
df_g2 = df.iloc[:, 5:10].copy()
df_g2.columns = ['Hora', 'T1_TCOR', 'T2_TCOR', 'T3_TCOR', 'T4_TCOR']
df_g2['Hora'] = df_g2['Hora'].str.extract(r'(\d{2}:\d{2}:\d{2}\.\d{3})')  # Extrae hora
df_g2.dropna(inplace=True)
df_g2['Tiempo_s'] = convertir_a_segundos(df_g2['Hora'])

# === Función de graficado ===
def graficar(df, titulo):
    plt.figure(figsize=(10, 6))
    plt.plot(df['Tiempo_s'], df['T1_TCOR'], label='T1_TCOR')
    plt.plot(df['Tiempo_s'], df['T2_TCOR'], label='T2_TCOR')
    plt.plot(df['Tiempo_s'], df['T3_TCOR'], label='T3_TCOR')
    plt.plot(df['Tiempo_s'], df['T4_TCOR'], label='T4_TCOR')
    plt.xlabel('Tiempo transcurrido [s]')
    plt.ylabel('Temperatura corregida [°C]')
    plt.title(titulo)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# === Graficar ambos grupos ===
graficar(df_g1, 'Grupo 1: Temperaturas vs Tiempo')
graficar(df_g2, 'Grupo 2: Temperaturas vs Tiempo')
```
