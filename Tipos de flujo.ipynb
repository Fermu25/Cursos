```python
import numpy as np
import matplotlib.pyplot as plt

# Tiempo simulado (0 a 10 segundos)
t = np.linspace(0, 10, 1000)

# Funciones que simulan el flujo vs. tiempo para diferentes bombas
def flujo_neumatico(t):
    return np.piecewise(t, [t % 1 < 0.1], [1, 0])

def flujo_jeringa(t):
    return 0.5 * t

def flujo_peristaltica(t):
    return 0.3 * np.sin(2 * np.pi * t) + 0.5

def flujo_diafragma(t):
    return 0.2 * np.sin(2 * np.pi * t) + 0.4

# Graficar los flujos
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(t, flujo_neumatico(t))
plt.title('Flujo vs Tiempo - Bomba Neumtica')
plt.xlabel('Tiempo (s)')
plt.ylabel('Flujo')

plt.subplot(2, 2, 2)
plt.plot(t, flujo_jeringa(t))
plt.title('Flujo vs Tiempo - Bomba de Jeringa')
plt.xlabel('Tiempo (s)')
plt.ylabel('Flujo')

plt.subplot(2, 2, 3)
plt.plot(t, flujo_peristaltica(t))
plt.title('Flujo vs Tiempo - Bomba Peristltica')
plt.xlabel('Tiempo (s)')
plt.ylabel('Flujo')

plt.subplot(2, 2, 4)
plt.plot(t, flujo_diafragma(t))
plt.title('Flujo vs Tiempo - Bomba de Diafragma')
plt.xlabel('Tiempo (s)')
plt.ylabel('Flujo')

plt.tight_layout()
plt.show()
```
