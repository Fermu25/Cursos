##Ejercicio 1. Diseño de un telescopio refractor


Un fabricante de telescopios está diseñando un telescopio refractor utilizando lentes convergentes. El telescopio constará de dos lentes: una lente objetivo de distancia focal $ f_o $ y una lente ocular de distancia focal $ f_e $. El objetivo del telescopio es producir una imagen ampliada y nítida de objetos lejanos.

Para calcular la magnificación $ M $ del telescopio y la distancia focal del mismo se tiene la siguiente
fórmula:

$$ M = \frac{f_e}{f_o} $$



Mientras que la distancia focal del sistema, denotada como $ f_s $, se expresa en terminos del lente objetivo y del lente ocular:

$$ \frac{1}{f_s} = \frac{1}{f_o} + \frac{1}{f_e} $$

$\textbf{a)}$ El actual diseño del telescopio tiene los siguientes valores, para la distancia focal de la lente objetivo $ f_o $ = 0.1 metros, y para la distancia focal de la lente ocular $ f_e $ = 0.02 metros. Calcula la magnificación del telescopio y la distancia focal del sistema.


$\textbf{b)}$ Como parte del trabajo el fabricante está probando diferentes distancias focales del sistema para obtener mejores magnificaciones. Realiza una gráfica de la magnificación en función de la distancia focal del sistema. Si la distancia focal del sistema toma valores entre  0.01 - 2 metros, que tu gráfica tenga almenos 100 puntos.  

$\textbf{c)}$ Discute **con tus propias palabras** el significado físico de la gráfica, y menciona cuál es la mejor distancia focal del sistema $ f_s $. Obten, para esta distancia focal del sistema, el valor de la distancia focal de la lente objetivo $ f_o $ .

La magnificación del telescopio $M$ en función de la distancia focal del sistema $ f_s $ se puede calcular como:

$$ M = \frac{f_e}{f_s} $$

Toma $ f_e = 0.02 m$ y calcula M para diferentes valores de $ f_s $

```python
import matplotlib.pyplot as plt
import numpy as np
import time
#Definición de las funciones
def dist_focal(fo,fe):
  y=(1/fo)+(1/fe)
  return 1/y

def magnificacion_objetivo(fe,fo):
  M1=fe/fo
  return M1
def magnificacion_sistema(fe,fs):
  M2= fe/fs -1
  return M2
t1=time.time()
#Parte a
objetivo=0.1
ocular=0.02
foco=dist_focal(objetivo,ocular)
mag=magnificacion_objetivo(ocular,objetivo)
print('La distancia focal del telescopio es: %.4f metros y la magnificación es: %.4f' %(foco,mag) )
t2=time.time()
print('Tiempo de ejecusión',t2-t1,'segundos','Tipo de tiempo lineal 6n')
#parte b
t3=time.time()
x=np.linspace(0.01,2,400)
plt.plot(x,magnificacion_sistema(ocular,x),'*r')
plt.xlabel('Distancia focal del sistema')
plt.ylabel('Magnificación')
t4=time.time()
print('Tiempo de ejecusión',t4-t3,'segundos','Tipo de tiempo polinomial 2n*400')
#parte c
#La gráfica indica que hay magnificación positiva en valores de la distancia focal del sistema cercanos a cero se verá una imagen derecha, pero estará reducida
#Se espera que entre mayor cercano sea a cero la distancias focal del sistema, la ampliacion sea mayor que 1, pero eso es muy dificil para un telescopio
#Por otro lado la magnificación es negativa mientras la distancia focal del sistema incrementa, es decir se verán imagenes reducidas pero invertidas y nuevamente
#Entre mayores sean las distancias focales del sistema, se espera que la amplificación invertida sea mayor que 1
#la mejor distancia focal del sistema es aquella donde se obtiene la mayor amplificación de imagen, dicha distancia focal del sistema se obtiene con distancias focales
#del objetivo y del ocular grandes lo cual si es posible en un telescopio, así que fs=2.
#Si fs=2, la distancia objetivo es:
def dist_focal_obj(fs,fe):
  z=(fs-fe)/(fs*fe)
  return 1/z
t5=time.time()
objetivo=dist_focal_obj(2,0.02)
print('La distancia focal del objetivo es:%.4f' %(objetivo))
t6=time.time()
print('Tiempo de ejecusión',t6-t5,'segundos','Tipo de tiempo lineal 3n')
```
