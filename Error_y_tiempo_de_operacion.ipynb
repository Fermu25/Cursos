Errorres en métodos numéricos
Para medir los errores numéricos usamos la nomenclatura de los errorres experimentales. Por ejemplo usamos la notación de dar el valor resultante al error $\pm$ al error. \\
$$h=h_{nume}\pm error$$
$$Ε_{error total}=Ε_{error total}+E_{APF}$$
APF:Aritmética de punto flotante


```python
from random import *
l=[]
for i in range(10):
  x=uniform(0.8,2.8) #val aleatorios en un intervalo
  l.append(x)

n=len(l)
suma=0
for i in range(n):
  suma+=l[i]
promedio = suma/n
error= max(l)-min(l)
print("El promedio es: \n", promedio,'+-', error)
```

```python
from math import *
def error_abs(real,aprox):
  return abs(real-aprox)
def error_rel(real,aprox):
  return (abs(real-aprox))/real
def error_por(real,aprox):
  return (abs(real-aprox)/real)*100

x_teo=e
x_aprox=19/7
e_abs=error_abs(x_teo,x_aprox)
e_rel=error_rel(x_teo,x_aprox)
e_por=error_por(x_teo,x_aprox)

print(e_abs)
```

Considera dos números
$$x=1$$
$$y=1+1\times 10^{-14} \sqrt(2)$$
En python calcula $\sqrt(2)$ cómo:
$$\sqrt(2)=1\times 10^{14} (y-x)$$

```python
#Underflow y Overflow, flotante más grande de python es 10**308 y 2**-1022 el más pequeño
from decimal import Decimal as D, getcontext as C
import pandas as pd
from math import *

C().prec=17
x=1
y=1+(1e-14)*sqrt(2)
def R1(x,y):
  return (1e14)*(y-x)
def R2(x,y):
  return D(1e14)*(D(y)-D(x))

print(R1(x,y),R2(x,y))
residuo=[]
idx=[]
for i in range(10):
  y=1+(10**(-i))*sqrt(2)
  idx.append(y)
  err=abs(float(R2(+D(x),+D(y)))-R1(x,y))
  residuo.append([R2(+D(x),+D(y)),R1(x,y), err*100])
cols=['Aprox','Real', '%']
pd.DataFrame(residuo, index=idx, columns=cols)
```
