```python
import numpy as np
import matplotlib.pyplot as plt

# Función
f = lambda x: x**3 - 10*x**2 + 5

# Método de aproximación de la secante modificado para graficar
def secante_grafico(x0, x1, epsilon=1e-8):
    contador = 0
    intervalo = abs(x1 - x0)
    # Preparar la gráfica de la función
    x = np.linspace(-2, 10, 400)
    plt.figure(figsize=(10, 8))
    plt.plot(x, f(x), label='f(x)')
    plt.title('Evolución de la recta secante')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.axhline(color='black', lw=0.5)
    plt.axvline(color='black', lw=0.5)

    while intervalo >= epsilon:
        if f(x1) - f(x0) == 0:
            print('La división por cero no está permitida')
            break
        xn = x1 - ((x1 - x0) / (f(x1) - f(x0))) * f(x1)
        # Graficar la recta secante
        secante_y = lambda x, x0, x1: (f(x1) - f(x0)) / (x1 - x0) * (x - x0) + f(x0)
        plt.plot(x, secante_y(x, x0, x1), label=f'Iteración {contador+1}')

        x0, x1 = x1, xn
        intervalo = abs(x1 - x0)
        contador += 1

    plt.legend()
    plt.show()
    return xn, contador

# Ejecución del método de la secante con gráficas
x0, x1 = 0, 1  # Puntos iniciales
epsilon = 1e-5  # Precisión deseada
raiz, iteraciones = secante_grafico(x0, x1, epsilon)
print(f"Raíz encontrada: {raiz} en {iteraciones} iteraciones")
```
