Método de Euler para ecuaciones diferenciales
$$
\frac{dx}{dt}=\frac{x^2}{t}
$$
$$
\int \frac{dx}{dt} = \int \frac{x^2}{t}
$$
$$
\frac{-1}{x}= ln(t) +c
$$
Pero la siguiente ecuacion no se resuelve analíticamente(método de euler es el método más arcaíco, pero es el más sencillo)
$$
\frac{dx}{dt}=\frac{2x}{t}+\frac{3x^2}{t^3}
$$
La forma general de una EDO es:
$$
\frac{dx}{dt}=f(x,t); f(x_1,x_2,x_3,...,x_n ,t)
$$
Se requieren condiciones inicales para programarlo, es decir $x(t_0)=x_0$
Ahora supongamos que tenemos x(t) para alguna t específica, hacemos Taylor:
$$
x(t+h)=x(t)+h\frac{dx}{dt}+\frac{1}{2} h^2 \frac{d^2 x}{dt}+....
$$
$$
x(t+h)=x(t)+h\cdot f(x_1,t)+O(h^2)
$$
Eso es Euler, se divide un intervalo en subintervalos homogéneos y se apróxima la función por intervalos utilizando el polinomio de Taylor asociado, se puede calcular cuanto valo la función en $a+h_1$ y sabiendo ese valor puedo calcular $a+h_2$ y así hasta cubrir todo el intervalo. Para el método de Euler se necesita conocer la función $f(x,t)$, el intervalo en el que se va a evaluar la función $[a,b]$ y el número de pasos $N$, de modo que $h=\frac{b-a}{N}$ y $N=\frac{b-a}{h}$

El error por paso de este método es cuadrático $O(h^2)$ el que acarrea Euler por cada paso, pero el error total es la suma del error de cada paso, se debe sumar entoces:
$$
\[
\sum_{k=0}^{N-1} (\frac{1}{2} h^2 \frac{d^2 x}{dt^2})=\frac{h}{2}\sum_{k=0}^{N-1} (h\frac{df}{dt})\approx \int_{a}^{b} \frac{h}{2}\frac{df}{dt}dt=\frac{h}{2}[f(x(b),b)-f(x(a),a)]
\]
$$
Suponiendo h pequeña, $t_k =a+kh$, $x_k=x(t_k)$ \\
Ejercicio:
Graficar $\frac{dx}{dt}=-x^3+sen(t)$ en el intervalo [0,10] con 1000 pasos y condiciones iniciales $x(t=0)=0$


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Definimos x:
t= np.linspace(0, 10, 1000)
#Definimos la función
def f(x,t):
  y=-x**3+np.sin(t)
  return y
#Definimos el intervalo y N
a=0
b=10
N=1000
h=(b-a)/N #Espaciamiento de los pasos
#Condiciones iniciales
x0=0

#método de Euler de forma iterativa, en cada punto se aproxima con rectas, ese es el significado geométrico
valores=[]
for i in t:
    x1=x0+h*f(x0,i)
    valores.append(x1)
    x0=x1

#Gráfico

plt.plot(t,valores)
```

$$
y^{\prime}+y=\frac{e^{-x} \tan(x)}{x}
$$
con y(1)=0 en el intervalo $[1,1.5]$ con N=1000

```python
from math import *
x=np.linspace(1,1.5,1000) #x representa al tiempo t y y a los valores en x, pero como aquí se usan x y y entonces se hizo así.
f=lambda y,x: -y+(np.exp(-x)*np.tan(x))/x
a=1
b=1.5
N=1000
h=(b-a)/N
y1=0
valores=[]
for i in x:
  y=y1+h*f(y1,i)
  valores.append(y)
  y=y1
plt.plot(x,valores)
```
